```python
# Import pandas terlebih dahulu
import pandas as pd

# buat dataframe dan simpan dengan nama "data" (penamaab dibebeaskan)
data = pd.DataFrame({'No'    : [1, 2, 3, 4, 5, 6], # ada 6 element
                     'Nama'  : ['Andi Gunawan', 'Beni Tri', 'Cinta Laudya', 'Deni Hermawan', 'Endang Gusti', 'Firman'],
                     'Alamat': ['Jakarta', 'Bogor', 'Bekasi', 'Bekasi', 'Aceh', ''], # ada 5 element
                     'Usia'  : [21, 24, 29, 24, 23, None] # ada 5 element
                    })

# run df
data
```

### c.2 Kesamaan SQL Condition

#### c.2.1 Filter Column in (a,b,c)

```python
df_customers[df_customers['city'].isin(['Sukabumi','Cirebon','Tasikmalaya','Singkawang'])]
```

Gunakan ini untuk mengambil dataset dengan kondisi __a__, __b__ dan __c__ <br>
dimana ke-3 nilai tersebut sudah exact

Konsep mirip dengan __WHERE COLUMN IN (a,b,c)__ pada SQL

```python
# CONTOH-1 : In case kita mengetahui nomor hp lengkap seperti berikut:

df_customers[df_customers['phone'].isin([6285244922, 6287840324, 6285249258])]
```

```python
# CONTOH-2 : In case kita mengetahu nama lengkap Ficky, Rianita dan Hilary

df_customers[df_customers['name'].isin(['Ficky Rizkyananta','Rianita Santi','Hilary Adha'])]
```

```python
# CONTOH-3 : TIDAK BISA DIGUNAKAN jika hanya mengetahui sebagian dari string, seperti kondisi berikut:

df_customers[df_customers['name'].isin(['Ficky','Rianita','Hilary'])]
```

#### c.2.2 Filter Column Null and Not Null

```python
df_customers[df_customers["phone"].notnull()]
```

### c.3 Multiple Condition

Digunakan saat melakukan filtering dengan banyak kondisi <br>

Konsep mirip dengan __WHERE COLUMN_A = XXXX and COLUMN_B IN (X,Y,Z) and COLUMN_C LIKE 'YY'__ pada SQL <br>
Melakukan filtering pada column A bernilai XXXX, column B bernilai X, Y dan Z dan column C yang mengandung kata YY

```python
from IPython import display
display.Image("assets/filter_multi_column.png", width=600)
```

```python
# CONTOH :

df_customers[ (df_customers['city'].isin(['Bogor','Pontianak'])) 
                 & (df_customers['active']==True) 
                 & (df_customers['age']>20)
                 & (df_customers['website_recency']<30)
                ]
```

```python
# Syarat:
## 1. income group ada di (1, 2)
## 2. tempat inggal ("Jakarta", "Bogor", "Depok", "Tangerang", "Bekasi")
df_customers[
    (df_customers["income_group"].isin([1, 2]))
    | (df_customers["city"].isin(["Jakarta", "Bogor", "Depok", "Tangerang", "Bekasi"]))
]
```
